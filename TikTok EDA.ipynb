## <b>3. Cleaning </b>

### Missing Values

From the discovering phase we analyzed there were missing values in the data set, Now we will work to remove those values

```python
# First we will analyze the rows that have claim_status as NaN since this is our key variable
df1 = df[df['claim_status'].isna()]
```

```python
df1
```

After analyzing the NaN data we can conclude that among 19382 only 298 rows have NaN as claim_status which makes only 1.53% of whole data moreover the rows that have null claim_status also does not have other important analysis variable like video_view_count, video_like_count, video_share_count, video_download_count.

### Outliers

First step is to identify if there are any outliers, How to deal with them comes second which will be done later

```python
print("Mean", np.mean(df['video_like_count']))
print("Median", np.median(df['video_like_count']))
```

The considerably high mean value relative to median shows that data is right skewed and the possibility of outliers are more on right side

Write a for loop that iterates over the column names of each count variable. For each iteration:
###### Calculate the IQR of the column
###### Calculate the median of the column
###### Calculate the outlier threshold (Q3 + 1.5 * IQR)
###### Calculate the numer of videos with a count in that column that exceeds the outlier threshold
###### Print "Number of outliers, {column name}: {outlier count}"

```python
#Identification of outliers
for x in df.select_dtypes(float).columns:
    Q1 = df[x].quantile(0.25)
    Median = df[x].quantile(0.50)
    Q3 = df[x].quantile(0.75)
    IQR = Q3 - Q1
    upperThreshold = Q3 + 1.5 * IQR
    outlierCount = df[df[x] > upperThreshold].shape[0]
    print("Number of outliers in " ,x ," : " ,outlierCount)
```

## <b>4. Analyzing</b> 

In this step we will analyze and try to prove the inital hypothesis made in discovering phase

##### 1. For videos with higher view and like count, claim_status is claim 

```python
df2 = df.sort_values(by=['video_view_count','video_like_count'],ascending=False)
df2.head(50)
```

```python
df2.tail(50)
```

##### <b>The hypothesis that videos with higher view and like count have claim_status 'claim' is proved</b>

##### 2. Mostly verified users posts lesser claims than not verified

```python
df['claim_status'].groupby(df['verified_status']).value_counts()
```

##### <b>This result speaks to the fact that most of the users of tiktok are not verified and veriifed users do post lesser claims than not verified users </b>

##### 3. Users with claims are more likely to be banned and less likely to be active as compared to those with opinion

```python
df3 = df.groupby(df['claim_status'])['author_ban_status'].value_counts()
df3
```

##### <b>People with opinion who are banned are only 3% and active are 93%  but people with claim who are banned are approximately 15% and the one who are active are 68% which proves that users with claims are more likely to be banned and less likely to be active as compared to those users with opinion</b>

##### 4. Video with more duration is more likely to have claim

```python
df4 = df.groupby(df['claim_status'])['video_duration_sec'].mean()
df4
```

##### <b>The hypothesis is rejeted as videos marked whether claim or opinion averages 32 second duration</b>
